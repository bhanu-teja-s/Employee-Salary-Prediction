```python
X = data[["Years","Job Rate"]]
y = data["Annual Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,predslr)
```

```python
import joblib
joblib.dump(lr,"linearmodel.pkl")
```

```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np

# Set page configuration
st.set_page_config(page_title="ðŸ’¼ Salary Prediction App", page_icon="ðŸ’°", layout="centered")

# Load model safely
@st.cache_resource
def load_model():
    try:
        return joblib.load("linearmodel.pkl")
    except FileNotFoundError:
        st.error("âŒ Model file not found. Please upload 'linearmodel.pkl'.")
        return None

model = load_model()

# App title and description
st.title("ðŸ’¼ Salary Prediction App")
st.markdown(
    """
    Welcome to the Salary Prediction App!
    ðŸŽ¯ Use this tool to **predict employee salaries** based on:
    - ðŸ¢ Years at company
    - ðŸ’¼ Job performance rating
    """
)

st.divider()

# Input form
with st.form("prediction_form"):
    years = st.number_input("ðŸ“… Years at Company", value=1, step=1, min_value=0)
    jobrate = st.number_input("â­ Job Performance Rating", value=3.5, step=0.5, min_value=0.0, max_value=5.0)
    submit = st.form_submit_button("ðŸ” Predict Salary")

st.divider()

# Prediction logic
if submit:
    if model is not None:
        X = np.array([[years, jobrate]])
        prediction = model.predict(X)[0]
        st.success(f"ðŸ’° Estimated Salary: **â‚¹{prediction:,.2f}**")
        st.balloons()
    else:
        st.warning("Model couldn't be loaded. Please check your file.")
```
